```python
import pandas as pd
import warnings
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, mean_squared_error
import numpy as np

# Load your data and preprocess it as you've shown in your code

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Encode the 'gender' column using one-hot encoding
X_train = pd.get_dummies(X_train, columns=['gender'], drop_first=True)
X_test = pd.get_dummies(X_test, columns=['gender'], drop_first=True)

# Initialize an empty DataFrame to store results
results = pd.DataFrame(columns=['K', 'Accuracy', 'RMSE'])

# Create a loop to iterate through different values of K (number of neighbors)
for k in range(1, 21):  # You can adjust the range as needed
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    results = results.append({'K': k, 'Accuracy': accuracy, 'RMSE': rmse}, ignore_index=True)

# Print the results
print(results)
```
