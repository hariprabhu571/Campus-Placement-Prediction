```python
X = data.drop('status',axis=1)
y= data['status']
```

### 10. Splitting The Dataset Into The Training Set And Test Set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

### 11. Import The models

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Encode the 'gender' column using one-hot encoding
X_train = pd.get_dummies(X_train, columns=['gender'], drop_first=True)
X_test = pd.get_dummies(X_test, columns=['gender'], drop_first=True)
```

### 12. Model Training

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred4 = dt.predict(X_test)
```

### 14. Evaluating the Algorithms

```python
from sklearn.metrics import accuracy_score
```

```python
score4=accuracy_score(y_test,y_pred4)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mse4 = mean_absolute_error(y_test,y_pred4)
print("mse ",mse4)
```

```python
print(score4)
```

```python
final_data = pd.DataFrame({'Models':['DT'],
            'ACC':[score4*100]})
```

```python
max_depth = dt.tree_.max_depth
print(f'Maximum depth of the Decision Tree model: {max_depth}')
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
accuracy_values = []
rmse_values = []
# Define a range of values for max_depth to experiment with
max_depth_values = range(1, 101)

for max_depth in max_depth_values:
    # Create and train the Decision Tree model with the specified max_depth
    dt = DecisionTreeClassifier(max_depth=max_depth)
    dt.fit(X_train, y_train)

    # Make predictions
    y_pred = dt.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_values.append(accuracy)

    # Calculate RMSE (Root Mean Squared Error)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    rmse_values.append(rmse)

    print(f"Epoch with max_depth = {max_depth}: Accuracy = {accuracy}, RMSE = {rmse}")
```
